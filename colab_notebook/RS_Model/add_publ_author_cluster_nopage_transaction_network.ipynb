## **1. Language**

```python
#first feature to be evaluated to find out about the similarities in the social network: language.
#here, we simply look at values of language column in transaction network dataset, to have the initial state.
#at the end, we look at this data again, to find out whether the result is different or not.
aut_type = trans_network.groupby('language',sort = False).agg({'itemID_i': 'count'}).reset_index()
aut_type.sort_values(by=['itemID_i'], ascending=False, inplace=True, ignore_index= True)
print(aut_type)
```

```python
def calculate_lang_difference(item_i, item_j, index):
    
    #Find item i&j s' index in dataframe df (item data)
    item_i_index = int(df[df['itemID'] == item_i].index.values)
    item_j_index = int(df[df['itemID'] == item_j].index.values)
    
    #1. Language: whether item i&j in same language (yes: 1 or no: 0)
    lang_i = df.iloc[item_i_index]['language']
    #print(lang_i)
    lang_j = df.iloc[item_j_index]['language']
    #print(lang_j)
    if (lang_i == lang_j):
        trans_network.loc[index,'language'] = 1
    else:
        trans_network.loc[index,'language'] = 0
```

```python
for index, i in trans_network.iterrows():
    if (index%1000 == 0):
        print(index)
    calculate_lang_difference(i['itemID_i'], i['itemID_j'], index)
```

```python
aut_type = trans_network.groupby('language',sort = False).agg({'itemID_i': 'count'}).reset_index()
aut_type.sort_values(by=['itemID_i'], ascending=False, inplace=True, ignore_index= True)
print(aut_type)
```

```python
#since we did not changed anything w.r.t titles and main topics, we do not have to revise the result.
#So, we can focus on new features. the first new feature is publisher clusters

def calculate_publisher_cluster_difference(item_i, item_j, index):
    
    #Find item i&j s' index in dataframe df (item data)
    item_i_index = int(df[df['itemID'] == item_i].index.values)
    item_j_index = int(df[df['itemID'] == item_j].index.values)
    
    #6. Publisher clusters. since we have different clusters for different languages: first, we capture the language of
    #both books. Then, if the languages are different, the similarity would be zero. But, if they are same, we should look
    #for each specific language. so, if both books are written in English, German, Spanish, Italian, or French, then, we
    #simply use the cosine similarity matrix that we extracted during the clustering. Otherwise, we set the measure to 1,
    #which means that for all infrequent languages, we basically use the same score equal to 1 for all books.
    
    pub_cluster_i = df.iloc[item_i_index]['publisher_cluster']
    pub_cluster_j = df.iloc[item_j_index]['publisher_cluster']
    
    lang_i = df.iloc[item_i_index]['language']
    lang_j = df.iloc[item_j_index]['language']
    
    if (lang_i != lang_j):
        trans_network.loc[index,'publisher_cluster'] = 0
    elif (lang_i == 'en'):
        trans_network.loc[index,'publisher_cluster'] = cosine_sim_pub_en [pub_cluster_i][pub_cluster_j]
    elif (lang_i == 'de'):
        trans_network.loc[index,'publisher_cluster'] = cosine_sim_pub_de [pub_cluster_i][pub_cluster_j]
    elif (lang_i == 'es'):
        trans_network.loc[index,'publisher_cluster'] = cosine_sim_pub_es [pub_cluster_i][pub_cluster_j]
    elif (lang_i == 'it'):
        trans_network.loc[index,'publisher_cluster'] = cosine_sim_pub_it [pub_cluster_i][pub_cluster_j]
    elif (lang_i == 'fr'):
        trans_network.loc[index,'publisher_cluster'] = cosine_sim_pub_fr [pub_cluster_i][pub_cluster_j]
    else:
        trans_network.loc[index,'publisher_cluster'] = 1
```

```python
#The second new feature is author clusters

def calculate_author_cluster_difference(item_i, item_j, index):
    
    #Find item i&j s' index in dataframe df (item data)
    item_i_index = int(df[df['itemID'] == item_i].index.values)
    item_j_index = int(df[df['itemID'] == item_j].index.values)
    
    #7. Author clusters. If at least one of the books does not have any author cluster, the similarity is considered to be zero.
    #otherwise, we use the cosine similarity measure that we have calculated before.
    
    author_cluster_i = df.iloc[item_i_index]['author_cluster']
    author_cluster_j = df.iloc[item_j_index]['author_cluster']
    
    if ((author_cluster_i == 20) and (author_cluster_j == 20)):
        #if two unknown meet each other, it is similar to p(coin = tails/head)
        trans_network.loc[index,'author_cluster'] = 0.5
    elif ((author_cluster_i != 20) and (author_cluster_j != 20)):  
        trans_network.loc[index,'author_cluster'] = cosine_sim_author [author_cluster_i][author_cluster_j]
    else:
        #if one unknown meet one known cluster, we again use p(coin = tails/head)
        trans_network.loc[index,'author_cluster'] = 0.5
```

```python
import random
```

```python
#The third new feature is number of pages

def calculate_num_pages_difference(item_i, item_j, index):
    
    #Find item i&j s' index in dataframe df (item data)
    item_i_index = int(df[df['itemID'] == item_i].index.values)
    item_j_index = int(df[df['itemID'] == item_j].index.values)
    
    #8. number of pages. If at least one of the books does not have number of pages. the similarity is considered to be 0.5,
    #just like tossing a coin. otherwise, we use the absolute difference of pages to define the similarity.

    num_pages_i = df.iloc[item_i_index]['number_pages']
    num_pages_j = df.iloc[item_j_index]['number_pages']
    
    if ((num_pages_i == 0) and (num_pages_j == 0)):
        #since number of NaN values is significant: if two unknown meet each other, it is similar to a random process
        #Thus, we generate a random number between 0 and 1 for these cases. 
        n = random.random()
        trans_network.loc[index,'number_pages'] = n
    elif ((num_pages_i != 0) and (num_pages_j != 0)):
        
        #if we check the disrtibution of number of pages, we can find out that the number of pages under 24, and over 640
        #is not significant. so, we don't want to let these outliers to affect our similarity measure.
        if (num_pages_i <= 24):
            num_pages_i = 24
        elif (num_pages_i >= 640):
            num_pages_i = 640
        if (num_pages_j <= 24):
            num_pages_j = 24
        elif (num_pages_j >= 640):
            num_pages_j = 640
        
        trans_network.loc[index,'number_pages'] = 1-(abs(num_pages_i-num_pages_j)/(640-24))
    else:
        #since number of NaN values is significant: if one unknown meet another book with known number of pages.
        #we assume that this is similar to a random process
        #Thus, we generate a random number between 0 and 1 for these cases.
        n = random.random()
        trans_network.loc[index,'number_pages'] = n
```

```python
for index, i in trans_network.iterrows():
    if (index%1000 == 0):
        print(index)
    calculate_num_pages_difference(i['itemID_i'], i['itemID_j'], index)
```

```python
aut_type = trans_network.groupby('number_pages',sort = False).agg({'itemID_i': 'count'}).reset_index()
aut_type.sort_values(by=['itemID_i'], ascending=False, inplace=True, ignore_index= True)
print(aut_type)
```

## **7. Description**

```python
import nltk
nltk.download('stopwords')
import spacy
import pickle
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
import re
import string
```

```python
# Function for removing NonAscii characters
def _removeNonAscii(s):
    return "".join(i for i in s if  ord(i)<128)

# Function for converting into lower case
def make_lower_case(text):
    return text.lower()

# Function for removing stop words
lang_dict = { "en" : 'english', "de" : 'german', "fr" : 'french', "es" : 'spanish', "it" : 'italian'}

def remove_stop_words(text, item_lang):
      language = lang_dict.get(item_lang, 'english')
      text = text.split()
      stops = set(stopwords.words(language)) #change to other language #"english"
      text = [w for w in text if not w in stops]
      text = " ".join(text)
      return text

# Function for removing punctuation
def remove_punctuation(text):
    tokenizer = RegexpTokenizer(r'\w+')
    text = tokenizer.tokenize(text)
    text = " ".join(text)
    return text

# Function for removing the html tags
def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)

# Stemming function
porter_stemmer = nltk.stem.porter.PorterStemmer()
def stem_words(words):
  return ' '.join([porter_stemmer.stem(word) for word in words.split()])

# Lemmatization function
nlp = spacy.load('en', disable=['parser', 'ner'])
def lemmatize_words(text):
  doc = nlp(text)
  spacy_lemmatized = " ".join([token.lemma_ for token in doc if str(token.lemma_) != '-PRON-'])
  return spacy_lemmatized
```

```python
df_temp = df[['description','language']]
df_temp = df_temp.rename(columns={'description':'processed_description'})
```

```python
def preprocess(text, language):
  text = _removeNonAscii(text)
  text = remove_punctuation(text)
  text = remove_html(text)
  text = make_lower_case(text)
  text = remove_stop_words(text, language)
  text = lemmatize_words(text)
  text = stem_words(text)
  return text

# stemming+lemmatization - takes 70min

for i in range(len(df_temp)):
    if i % 100 == 0:
      print(i)
    if ((pd.isnull(df_temp.loc[i, 'processed_description'])) or (df_temp.loc[i, 'processed_description'] == None)):
        df_temp.loc[i, 'processed_description'] = ''
    description = df_temp.at[i,'processed_description']
    language = df_temp.at[i, 'language']
    processesed_description = preprocess(description, language)
    df_temp.loc[i, 'processed_description'] = processesed_description
```
